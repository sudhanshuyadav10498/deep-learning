```python
import tensorflow as tf
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
(train_x, train_y), (test_x, test_y) = tf.keras.datasets.boston_housing.load_data()
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', input_shape=[train_x.shape[1]]),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
```

```python
model.summary()
```

```python
history = model.fit(train_x, train_y, epochs=100, validation_split=0.1)
```

```python
plt.plot(history.history['loss'], label="Training Loss")
plt.plot(history.history['val_loss'], label = 'Validation Los')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Epoch vs Loss")
plt.legend()
plt.plot()
```

```python
plt.plot(history.history['mse'], label="Training MSE")
plt.plot(history.history['val_mse'], label = 'Validation MSE')
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.title("Epoch vs MSE")
plt.legend()
plt.plot()
```

```python
plt.plot(history.history['mae'], label="Training MAE")
plt.plot(history.history['val_mae'], label = 'Validation MAE')
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.title("Epoch vs MAE")
plt.legend()
plt.plot()
```
